## CONNECT DB

```python
import pandas as pd
import sqlite3
import seaborn as sns
```

```python
con = sqlite3.connect("olist.db")
cur = con.cursor()
```

```python
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cur.fetchall())
```

## READ TABLE

```python
#read table product

df_olist_product = pd.read_sql("SELECT * FROM olist_products_dataset", con)

df_olist_product
```

```python
df_olist_product.dtypes
```

```python
#read table seller

df_olist_seller = pd.read_sql("SELECT * FROM olist_sellers_dataset", con)

df_olist_seller
```

```python
df_olist_seller.dtypes
```

```python
#read table order items

df_olist_order_items = pd.read_sql("SELECT * FROM olist_order_items_dataset", con)

df_olist_order_items
```

```python
df_olist_order_items.dtypes
```

```python
#read table product translation

df_product_translation = pd.read_sql("SELECT * FROM product_category_name_translation", con)

df_product_translation
```

```python
#read table olist orders
df_olist_orders = pd.read_sql("SELECT * FROM olist_order_dataset", con)
df_olist_orders
```

## COMBINE AND FILTER TABLE

#### DATA PRODUK

```python
#Menggabungkan tabel data produk dengan translasi nama produk berbahasa inggris
data_produk = pd.merge(df_olist_product, df_product_translation, on = "product_category_name", how = "left")
data_produk
```

```python
#filter kolom tabel yang akan digunakan
product_index_columns = [1, 11, 6, 7, 9, 8]
data_produk = data_produk.iloc[0:, product_index_columns]

#Mengubah header tabel
data_produk.columns = ["Product ID", "Nama Produk", "Berat (gr)", "Panjang (cm)", "Lebar (cm)", "Tinggi (cm)"]
data_produk
```

#### CUSTOMER

```python
#filter kolom tabel yang akan digunakan
customer_index_columns = [1, 4, 5]
customer = df_olist_order_customer.iloc[0:, customer_index_columns]

#Mengubah header tabel
customer.columns = ["Customer ID", "Kota", "Kode Provinsi"]
customer
```

#### SELLER

```python
#filter kolom tabel yang akan digunakan
seller_index_columns = [1, 3, 4]
seller = df_olist_seller.iloc[0:, seller_index_columns]

#Mengubah header tabel
seller.columns = ["Seller ID", "Kota", "Kode Provinsi"]
seller
```

#### ORDER

```python
#Menggabungkan tabel order dengan order items untuk mendapatkan kolom customer id
order = pd.merge(df_olist_order_items, df_olist_orders, on = "order_id", how = "left")
order
```

```python
#filter kolom tabel yang akan digunakan
order_index_columns = [1, 3, 4, 9, 6, 7]
order = order.iloc[0:, order_index_columns]

#Mengubah header tabel
order.columns = ["Order ID", "Product ID", "Seller ID", "Customer ID", "Harga Produk", "Ongkos Kirim"]
order
```

## DATA CLEANSING

#### MISSING VALUE

##### Data Produk

```python
#Menghitung persentase missing value
data_produk.isna().sum() / len(data_produk) * 100
```

```python
#Cek tipe data
data_produk.info()
```

```python
#imputasi data yg bertipe numerik ke mediannya
data_produk["Berat (gr)"] = data_produk["Berat (gr)"].fillna(data_produk["Berat (gr)"].median())
data_produk["Panjang (cm)"] = data_produk["Panjang (cm)"].fillna(data_produk["Panjang (cm)"].median())
data_produk["Lebar (cm)"] = data_produk["Lebar (cm)"].fillna(data_produk["Lebar (cm)"].median())
data_produk["Tinggi (cm)"] = data_produk["Tinggi (cm)"].fillna(data_produk["Tinggi (cm)"].median())

#imputasi data yg bertipe object ke "unknown"
data_produk["Nama Produk"] = data_produk["Nama Produk"].fillna("unknown")
```

```python
#casting tipe data
data_produk["Berat (gr)"] = data_produk["Berat (gr)"].astype("int64")
data_produk["Panjang (cm)"] = data_produk["Panjang (cm)"].astype("int64")
data_produk["Lebar (cm)"] = data_produk["Lebar (cm)"].astype("int64")
data_produk["Tinggi (cm)"] = data_produk["Tinggi (cm)"].astype("int64")
```
